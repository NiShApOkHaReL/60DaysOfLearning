```python
from sklearn.model_selection import train_test_split
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
# Prepare the data
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32).view(-1, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.2, random_state=2 )
```

```python
class DropOut(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(X_train.shape[1],32)
        self.l2 = nn.Linear(32,16)
        self.l3 = nn.Linear(16,1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        
    def forward(self,x):
        x = self.relu(self.l1(x))
        x = self.relu(self.l2(x))
        x = self.sigmoid(self.l3(x))
        
        return x
    
model = DropOut()    
```

```python
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model.parameters(),lr=0.01)
```

```python
epochs = 100
for i in range(epochs):
    #Training phase
    model.train()
    y_pred = model(X_train)
    loss = loss_fn(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    #Testing phase
    model.eval()
    with torch.no_grad():
        y_test_pred = model(X_test)
        test_loss = loss_fn(y_test_pred, y_test)
        
    print(f"Epoch [{i+1}/{epochs}], Train loss: {loss.item():.4f}, Test_loss: {test_loss.item():.4f}")
```

```python
class DropOut(nn.Module):
    def __init__(self):
        super().__init__()
        self.l1 = nn.Linear(X_train.shape[1],32)
        self.l2 = nn.Linear(32,16)
        self.l3 = nn.Linear(16,1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()
        self.dropout = nn.Dropout(0.5)
        
    def forward(self,x):
        x = self.relu(self.l1(x))
        x = self.dropout(x)  # Apply dropout after the first hidden layer
        x = self.relu(self.l2(x))
        x = self.dropout(x)  # Apply dropout after the second hidden layer
        x = self.sigmoid(self.l3(x))
        
        return x
    
model2 = DropOut()    
```

```python
loss_fn = nn.BCELoss()
optimizer = optim.Adam(model2.parameters(),lr=0.01)
```

```python
epochs = 100
for i in range(epochs):
    #Training phase
    model2.train()
    y_pred = model2(X_train)
    loss = loss_fn(y_pred, y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    #Testing phase
    model2.eval()
    with torch.no_grad():
        y_test_pred = model2(X_test)
        test_loss = loss_fn(y_test_pred, y_test)
        
    print(f"Epoch [{i+1}/{epochs}], Train loss: {loss.item():.4f}, Test_loss: {test_loss.item():.4f}")
```
